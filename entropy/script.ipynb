```python
import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 
                            'xtick.labelsize':20, 
                            'ytick.labelsize':20, 
                            'axes.labelsize':20, 
                            'figure.figsize':(12 * 2/3, 9 * 2/3)})
```

```python
import scipy, scipy.stats
```

```python
n = 2
probs = [1/n, 1/n]
h = scipy.stats.entropy(probs, base=2)
print(h)
```

```python
# https://en.wikipedia.org/wiki/Entropy_(information_theory)#Example
probs = [0.3, 0.7]
h = scipy.stats.entropy(probs, base=2)
print(h)
```

```python
probs = [1/3, 1/3, 1/3]
h = scipy.stats.entropy(probs, base=2)
print(h)
```

```python
for n in range(1, 5, 1):

  number_of_states = 3**n
  probs = [1/number_of_states for i in range(number_of_states)]
  h = scipy.stats.entropy(probs, base=2)
  
  print(n, number_of_states)
  print(probs)
  print()
```

```python
entropies = []; dimensions = []
for n in range(1, 5, 1):

  number_of_states = 3**n
  probs = [1/number_of_states for i in range(number_of_states)]
  h = scipy.stats.entropy(probs, base=2)
  print(n, h)
  
  entropies.append(h); dimensions.append(n)
```

```python
matplotlib.pyplot.plot(dimensions, entropies, 'o-', color='black')

matplotlib.pyplot.xlabel('Number of dimensions')
matplotlib.pyplot.ylabel('Entropy (bits)')
matplotlib.pyplot.xticks([1, 2, 3, 4])
matplotlib.pyplot.grid(ls=':')
matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```
