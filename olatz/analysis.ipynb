```python
import pandas, numpy, seaborn
```

```python
import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 
                            'font.family':'sans-serif', 
                            'xtick.labelsize':16, 
                            'ytick.labelsize':16, 
                            'figure.figsize':(16*(2/3), 9*(2/3)), 
                            'axes.labelsize':20
                           })
```

```python
input_file = 'Feno_Ornitho_LC_ADRIANO.xlsx'
```

```python
%%time
df = pandas.read_excel(input_file)
```

```python
unique_species = list(set(df['Nombre cient√≠fico']))
unique_species.sort()
print(len(unique_species), unique_species[:10])
```

```python
for species in unique_species[20:30]:
    
    print(species)
    species_specific_df = df[df['Nombre cient√≠fico'] == species]

    
    years_measured = list(set(species_specific_df['A√±o']))
    years_measured.sort()
    print(years_measured)

    # making analysis and figure
    x=[]; y=[]
    for month_measured in range(1, 12+1):
        print('\t', 'month', month_measured)
        medians_across_years = []
        for year_measured in years_measured:
            final_slice = species_specific_df[(species_specific_df['A√±o'] == year_measured) & (species_specific_df['Mes'] == month_measured)]
            values = final_slice['N√∫mero'].values
            observed_median = numpy.median(values)
            #observed_median = 12
            print('\t\t', 'year', year_measured, '#', len(values), values[:10], observed_median) 
            if numpy.isnan(observed_median) == False: 
                medians_across_years.append(observed_median)
    
        # average over years
        for element in medians_across_years:
            y.append(element); x.append(month_measured)
        
    # create figure
    print(x)
    print(y)
    seaborn.stripplot(x=x, y=y, color='black', size=2.5, native_scale=True)
    try:
        seaborn.regplot(x=x, y=y, order=2, x_estimator=numpy.median, color='skyblue')
        seaborn.regplot(x=x, y=y, order=3, x_estimator=numpy.median, color='gold')
    except:
        print('regression failed')

        
    matplotlib.pyplot.xlim([0.5, 12.5])
    matplotlib.pyplot.xticks(range(1, 12+1), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
    matplotlib.pyplot.xlabel('Month')
    matplotlib.pyplot.ylabel('Observed')
    matplotlib.pyplot.grid(ls=':')
    matplotlib.pyplot.title(species)
    
    matplotlib.pyplot.tight_layout()
    matplotlib.pyplot.show()

    

    print()
```
