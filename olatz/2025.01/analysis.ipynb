```python
print(vg.shape)

locations = vg['UTM']
locations = list(set(locations))

extra = 0
for location in locations:
    print(location)
    
    old_location = location.replace('30T', '')
    
    subset_cuan = cuan[cuan['UTM'] == old_location]
    subset_vg = vg[vg['UTM'] == location]

    species_vg = subset_vg['Species observadas'].to_list()
    species_cuan = subset_cuan['Species observadas'].to_list()

    #print(species_vg)
    #print(species_cuan)

    print('\t species in vg, species in cuan', len(species_vg), len(species_cuan))

    for species in species_cuan:
        if species not in species_vg:
            print('\t\t', 'adding from cuan', species, 'on', old_location)
            new_record = subset_cuan[subset_cuan['Species observadas'] == species]
            extra = extra + 1
            vg = pandas.concat([new_record, vg])
        else:
            print('\t\t\t', 'not adding from cual', species, 'on', old_location)

# checking on size
print(vg.shape, extra)
print(vg.shape[0] - extra)
```
